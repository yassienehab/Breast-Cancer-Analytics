**This is SEER Breast Cancer Data set from NCI program , which provides information on population-based cancer statistics. The dataset involved female patients with infiltrating duct and lobular carcinoma breast cancer diagnosed in 2006-2010***

#columns explination

By getting back to [data scource](https://ieee-dataport.org/open-access/seer-breast-cancer-data)

**1) Age** -> womwn's age

**2) Race** -> ethnicity (black-white-asin-unknown)

**3) marital status** ->(Single-Married-Separated-Divorced-Widowed)

**4) T Stage** -> T stands for Tumor size and extent — specifically, how large the tumor is and whether it has grown into nearby tissues.
- T1	Tumor ≤ 2 cm in greatest dimension
- T2	Tumor > 2 cm but ≤ 5 cm
- T3	Tumor > 5 cm
- T4	Tumor of any size that invades the chest wall or skin (more advanced local spread)

**5) N Stage** -> This stage describes whether cancer has spread to nearby regional lymph nodes, and if so, how many and where.


- N1	Cancer has spread to 1–3 axillary (underarm) lymph nodes
- N2	Cancer in 4–9 axillary lymph nodes, or spread to internal mammary nodes without axillary involvement
- N3	More advanced node involvement

**6) 6th Stage** ->
The "6th Stage" column refers to the AJCC 6th Edition Cancer Staging System, which was used by SEER to classify cancer severity based on the TNM system:

T = Tumor size and extent

N = Node involvement (lymph nodes)

M = Metastasis (spread to distant organs)

- (IIA)	Tumor 2–5 cm, no/mild lymph node involvement
- (IIB)	Tumor >2 cm with moderate lymph node involvement, or tumor >5 cm with no nodes
- (IIIA)	Larger tumor + significant lymph node spread (e.g., matted nodes or nodes fixed to other structures)
- (IIIB)	Tumor has grown into chest wall or skin, may have lymph node spread
- (IIIC)	Extensive lymph node involvement (e.g., underarm and collarbone nodes)

**7) Grade** -> level of differentiation
- Grade I (grade i - grade 1) -> well differentiated
- Grade II (grade ii - grade 2) -> moderately differentiate - intermediate differentiation
- Grade III (grade iii - grade 3) -> poorly differentiated
- Grade IV (grade iv - grade 4) -> undifferentiated - anaplastic

**8) A Stage** ->
- Regional —> A neoplasm that has extended

  1) beyond the limits of the organ of origin directly into surrounding organs or tissues

  2) into regional lymph nodes by way of the lymphatic system

  3) by a combination of extension and regional lymph nodes.

- Distant —> A neoplasm that has spread to parts of the body remote from the primary

  1)tumor either by direct extension

  2) by discontinuous metastasis (e.g., implantation or seeding) to distant organs, issues, or via the lymphatic system to distant lymph nodes.

**9) Tumor Size** -> the size of tumor in millmeters (mm)

**10) Estrogen Status** ->
  - positive -> The tumor has estrogen receptors (ER+) — hormone therapy may help slow or stop the cancer.
  - negative -> The tumor does NOT have estrogen receptors (ER−) — hormone therapy will not be effective.

**11) Progesterone Status** ->
  - Positive -> The tumor has progesterone receptors (PR+), and may respond to hormone therapy.
  - Negative -> The tumor has no progesterone receptors (PR−), so hormone therapy is less effective.

**12) Regional Node Examined** -> refers to the number of regional lymph nodes that were surgically removed and examined to check for the spread of cancer.(checl if lymph node has cancer or not)
  - Lymph nodes are a key pathway for cancer spread.

  -The more nodes examined, the more accurate the N Stage determination is.

**13) Reginol Node Positive** -> The number of those examined nodes that were found to contain cancer.

**14) Survival Months** -> represents the number of months a patient lived after being diagnosed with breast cancer

**15) Status** -> patient Alive or Dead

#Dealing with outliers

**Tumor size**

Q1 (25%) = 1.6

Q3 (75%) = 3.8

IQR = Q3 - Q1 = 2.2

Upper bound for outliers = Q3 + 1.5 * IQR = 3.8 + 1.5 * 2.2 = 7.1

" > 7" -> outlier

```python
bins = [0, 2, 5, 7, 15]
labels = ['Small', 'Medium', 'Large', 'Very Large']
breast_cancer_df['Tumor Size Group'] = pd.cut(breast_cancer_df['Tumor Size'], bins=bins, labels=labels)
```

- Age

- Divide age to categories and know how much patients in each category

```python
def age_to_age_group(age):
  if age>=30 and age<40:
    return "thirties"
  elif age>=40 and age<50:
    return "forties"
  elif age>=50 and age<60:
    return "fifties"
  else:
    return "sixties"
breast_cancer_df['Age_Stage']=breast_cancer_df['Age'].apply(age_to_age_group)
breast_cancer_df.head()
```

```python
fig,ax=plt.subplots()
age_order = ['thirties', 'forties', 'fifties', 'sixties']
sns.countplot(x='Age_Stage',data=breast_cancer_df,hue='Age_Stage',order=age_order,palette='Set2')
ax.set_xlabel("Patient's Age_Stage")
ax.set_ylabel("No of observations")
ax.set_title("How much patient in each category?")
```

- see how much people get live and how much people dead in each stage

```python
counts=breast_cancer_df[['Age_Stage','Status']].value_counts().unstack()
counts
fig,ax=plt.subplots()

counts = counts.reindex(age_order)

labels = counts.index

alive = counts['Alive']

dead = counts['Dead']

x = np.arange(len(labels))

width = 0.35

ax.bar(x - width/2, alive, width, label='Alive', color='green')

ax.bar(x + width/2, dead, width, label='Dead', color='red')

ax.set_xlabel('Age Group')

ax.set_ylabel('Number of Patients')

ax.set_title('Survival Status by Age Group')

ax.set_xticks(x)
ax.set_xticklabels(labels)

ax.legend()

ax.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
